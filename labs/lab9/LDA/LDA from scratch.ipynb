## Understanding Linear Discriminant Analysis (LDA)

Linear Discriminant Analysis (LDA) and Principal Component Analysis (PCA) are both techniques used for dimensionality reduction, but they approach the problem with different goals in mind:

- **Goal Alignment:** Like PCA, LDA aims to find the most effective dimensions for data classification. However, LDA's objective is more nuanced.
- **Class Separation:** LDA seeks dimensions that not only separate data but do so by maximizing the distance between class means while also minimizing the variance within each class when projected onto these dimensions. This dual focus distinguishes LDA from PCA.
- **Mathematical Framework:**
  - **Between-Class Scatter (Sb):** LDA computes the scatter matrix that quantifies the dispersion between different classes.
  - **Within-Class Scatter (Sw):** It also calculates the total within-class scatter matrix, capturing the variance within each class.
  - **Optimization:** The core of LDA involves solving for the eigenvectors of the expression inv(Sw)*Sb. This equation represents the mathematical solution to LDAâ€™s objective of maximizing between-class separation while minimizing within-class scatter.
- **Data Projection:** Finally, LDA projects the data points into a new space defined by these eigenvectors, effectively transforming the dataset into a form where classes are more distinctly separated.

This structured methodology, LDA not only enhances class separability but also provides a solid foundation for building more accurate classification models.

```python
# Imports
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
```

```python
class LDA:
    def __init__(self, n_components):
        self.n_components = n_components
        self.linear_discriminants = None

    def fit(self, X, y):
        n_features = X.shape[1]
        class_labels = np.unique(y)

        # Within class scatter matrix:
        # SW = sum((X_c - mean_X_c)^2 )

        # Between class scatter:
        # SB = sum( n_c * (mean_X_c - mean_overall)^2 )

        mean_overall = np.mean(X, axis=0)
        SW = np.zeros((n_features, n_features))
        SB = np.zeros((n_features, n_features))
        for c in class_labels:
            X_c = X[y == c]
            mean_c = np.mean(X_c, axis=0)
            # (4, n_c) * (n_c, 4) = (4,4) -> transpose
            SW += (X_c - mean_c).T.dot((X_c - mean_c))

            # (4, 1) * (1, 4) = (4,4) -> reshape
            n_c = X_c.shape[0]
            mean_diff = (mean_c - mean_overall).reshape(n_features, 1)
            SB += n_c * (mean_diff).dot(mean_diff.T)

        # Determine SW^-1 * SB
        A = np.linalg.inv(SW).dot(SB)
        # Get eigenvalues and eigenvectors of SW^-1 * SB
        eigenvalues, eigenvectors = np.linalg.eig(A)
        # -> eigenvector v = [:,i] column vector, transpose for easier calculations
        # sort eigenvalues high to low
        eigenvectors = eigenvectors.T
        idxs = np.argsort(abs(eigenvalues))[::-1]
        eigenvalues = eigenvalues[idxs]
        eigenvectors = eigenvectors[idxs]
        # store first n eigenvectors
        self.linear_discriminants = eigenvectors[0 : self.n_components]

    def transform(self, X):
        # project data
        return np.dot(X, self.linear_discriminants.T)
```

```python
data = datasets.load_iris()
X, y = data.data, data.target

# Project the data onto the 2 primary linear discriminants
lda = LDA(2)
lda.fit(X, y)
X_projected = lda.transform(X)

print("Shape of X:", X.shape)
print("Shape of transformed X:", X_projected.shape)

x1, x2 = X_projected[:, 0], X_projected[:, 1]

plt.scatter(
    x1, x2, c=y, edgecolor="none", alpha=0.8, cmap=plt.cm.get_cmap("viridis", 3)
)

plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.colorbar()
plt.show()
```
